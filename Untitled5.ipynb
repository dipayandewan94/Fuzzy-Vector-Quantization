```python
def uniform_midtread_quantizer(x, Q):
    '''
    x: input array
    Q: quantization step
    '''
    # limiter
    x = np.copy(x)
    idx = np.where(np.abs(x) >= 1)
    x[idx] = np.ones(x[idx].shape)
    # linear uniform quantization
    xQ = np.round(Q * np.floor(x/Q + 1/2),6)

    return xQ
```
